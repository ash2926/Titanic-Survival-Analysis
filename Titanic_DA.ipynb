```python
import pandas as pd
```

```python
#dropping columns which are not important and not numerical

cols_to_drop=['Name','Ticket','Cabin']

titanic =titanic.drop(cols_to_drop,axis = 1)
titanic.info()
```

```python
# missing values
# there are still missings values in age,embarked and there are object values in sex and embarked
# embarked and sex are imp colms so we cant drop them.
# so we convert these colms to individual colms whic have numerical data in the form of 0's and 1's
# creating a dummy variable.
dummies =[]
cols = ['Sex','Embarked']
for col in cols:
  dummies.append(pd.get_dummies(titanic[col]))
```

```python
# concat
titanic_dummies = pd.concat(dummies,axis=1)
```

```python
# here the sex values male and female are turned into columns and they have numerical data
# same for embarked as well, there are seperate columns for C,Q and S
titanic_dummies
```

```python
#concatenating new columns
titanic = pd.concat((titanic,titanic_dummies),axis =1 )
titanic
```

```python
# dropping duplicate columns
titanic = titanic.drop(['Sex','Embarked'],axis=1)
titanic
```

```python
titanic.info()
```

```python
#here age is the only column which has missing values
titanic['Age'].interpolate(inplace=True)
```
